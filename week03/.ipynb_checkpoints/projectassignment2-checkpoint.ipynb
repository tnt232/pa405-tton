This notebook explores spatial patterns in 311 calls related to homelessness, illegal dumping, and bulky item pickup in LA County. These public service calls are analyzed per capita to identify hotspots and areas of high need.

```python
#2. Load and Prepare Data
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx

# Load tract shapefile
gdf_tracts = gpd.read_file("data/LACountyTracts24.zip")

# Load 311 call data
df_calls = pd.read_csv("data/All_Years_311_Calls_Type_Calculations.csv")  # adjust path as needed

# Clean and prepare GEOID
df_calls['geoid_stripped'] = df_calls['geoid'].str.replace('^1400000US', '', regex=True)

# Merge on GEOID
gdf_merged = gdf_tracts.merge(df_calls, left_on='ID', right_on='geoid_stripped')

# Project to Web Mercator
gdf_merged = gdf_merged.to_crs(epsg=3857)

```

```python
#3. Map 1: Homeless Encampment Rate
fig, ax = plt.subplots(figsize=(12, 10))
gdf_merged.plot(column='homeless_encampment_rate', cmap='Reds', legend=True,
                edgecolor='white', linewidth=0.2, ax=ax)
ctx.add_basemap(ax, source=ctx.providers.CartoDB.Positron)
ax.set_title("Homeless Encampment Calls per 1,000 People", fontsize=15)
ax.axis('off')
plt.show()
```

```python
#4. Map 2: Illegal Dumping Rate
fig, ax = plt.subplots(figsize=(12, 10))
gdf_merged.plot(column='illegal_dumping_rate', cmap='Purples', legend=True,
                edgecolor='white', linewidth=0.2, ax=ax)
ctx.add_basemap(ax, source=ctx.providers.CartoDB.Positron)
ax.set_title("Illegal Dumping Calls per 1,000 People", fontsize=15)
ax.axis('off')
plt.show()
```

```python
#5. Map 3: Bulky Item Pickup Rate
fig, ax = plt.subplots(figsize=(12, 10))
gdf_merged.plot(column='bulky_items_rate', cmap='Blues', legend=True,
                edgecolor='white', linewidth=0.2, ax=ax)
ctx.add_basemap(ax, source=ctx.providers.CartoDB.Positron)
ax.set_title("Bulky Item Pickup Calls per 1,000 People", fontsize=15)
ax.axis('off')
plt.show()
```

```python
#6. Bar Chart: Top 10 Tracts by Homeless Calls per Capita
top10 = gdf_merged.sort_values(by='homeless_encampment_rate', ascending=False).head(10)
top10.plot.bar(x='ID', y='homeless_encampment_rate', legend=False, figsize=(10, 6))
plt.title("Top 10 Tracts by Homeless Encampment Call Rate")
plt.ylabel("Calls per 1,000 Residents")
plt.xlabel("Census Tract ID")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Analysis Summary

The maps and charts highlight geographic disparities in service calls across Los Angeles County:

- **Homeless encampment rates** are highest in central LA and the San Fernando Valley.
- **Illegal dumping** follows similar patterns but is more dispersed.
- **Bulky item pickup calls** are especially high in densely populated residential neighborhoods.

These patterns align with known inequities in housing, population density, and city service access. The spatial analysis reinforces the need for targeted interventions in high-need areas.
